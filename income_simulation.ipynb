```python
import pandas as pd
import random
import matplotlib.pyplot as plt

# Define courier IDs and probabilities
couriers = list(range(1, 201))  # 1-100 are one category, 101-200 are another category
probabilities = [0.5 if 1 <= courier <= 100 else 0.25 for courier in couriers]
```

```python
# Create DataFrame for couriers and probabilities
courier_df = pd.DataFrame({
    'Courier_ID': couriers,
    'Probability': probabilities
})

courier_df.head()
```

```python
# Define the number of orders per income tier
orders_per_tier = 500  # Number of orders per income tier
order_values = [50, 40, 30]  # Three income tiers
```

```python
# Function to simulate order assignment
def assign_order():
    return random.choices(couriers, probabilities)[0]

# Generate 500 orders for each income tier, totaling 1500 orders
orders = []
for value in order_values:
    for _ in range(orders_per_tier):
        orders.append({
            'Order_ID': len(orders) + 1,
            'Courier_ID': assign_order(),
            'Order_Value': value
        })
order_df = pd.DataFrame(orders)
print("\nOrder Assignment Results DataFrame:")
order_df 
```

```python
# Calculate total income for each courier
courier_income = order_df.groupby('Courier_ID')['Order_Value'].sum().reset_index()
courier_income 
```

```python
# Add a new column to distinguish full-time and part-time couriers
courier_income['Type'] = courier_income['Courier_ID'].apply(lambda x: 'Full-time' if x <= 100 else 'Part-time')
courier_income 
```

```python
# Set colors: blue for full-time couriers, green for part-time couriers
colors = courier_income['Type'].map({'Full-time': 'blue', 'Part-time': 'green'})
# Plot total income per courier with different colors for full-time and part-time couriers
plt.figure(figsize=(12, 6))  # Set figure size
plt.bar(courier_income['Courier_ID'], courier_income['Order_Value'], color=colors)  # Generate bar chart with color distinction
plt.xlabel('Courier ID')  # Set x-axis label
plt.ylabel('Total Income')  # Set y-axis label
plt.title('Total Income per Courier (Full-time vs Part-time)')  # Set title

# Add legend
plt.legend(handles=[plt.Rectangle((0,0),1,1,color='blue', label='Full-time'),
                    plt.Rectangle((0,0),1,1,color='green', label='Part-time')])

plt.show()  # Show plot
```

```python
# Sort by income in descending order
courier_income_sorted = courier_income.sort_values(by='Order_Value', ascending=False)

# Plot sorted income bar chart
plt.figure(figsize=(12, 6))  # Set figure size
plt.bar(range(len(courier_income_sorted)), courier_income_sorted['Order_Value'], color=courier_income_sorted['Type'].map({'Full-time': 'blue', 'Part-time': 'green'}))
plt.xlabel('Rank (by Income)')  # Set x-axis label
plt.ylabel('Total Income')  # Set y-axis label
plt.title('Total Income per Courier (Sorted by Income)')  # Set title

# Add legend
plt.legend(handles=[plt.Rectangle((0,0),1,1,color='blue', label='Full-time'),
                    plt.Rectangle((0,0),1,1,color='green', label='Part-time')])

plt.show()  # Show plot
```
